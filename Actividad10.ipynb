# **Analisis de los modelos del parque automotor de Boyaca.**

La media es 1986, lo cual indica que en promedio los automoviles en boyaca tiene  más de 30 años , esto puede acarrear varias consecuencias. En primer lugar, los coches más viejos pueden requerir reparaciones y mantenimiento más frecuentes, lo que puede resultar costoso. En segundo lugar, mientras que la compra de un coche nuevo puede conllevar una depreciación significativa en cuanto sale del concesionario, los coches más antiguos también pueden tener un valor de reventa más bajo. En tercer lugar, los coches más antiguos pueden no tener las mismas características de seguridad que los modelos más nuevos, lo que puede aumentar el riesgo de accidentes. En cuarto lugar, los coches más viejos pueden no ser tan eficientes en el consumo de combustible como los modelos más nuevos, lo que puede provocar un aumento de las emisiones y un mayor impacto en el medio ambiente. Por último, a medida que los coches envejecen, puede resultar más difícil encontrar piezas de repuesto cuando se necesitan reparaciones. Es importante señalar que estas consecuencias pueden variar en función de la marca y el modelo concretos del coche, así como de su estado y su historial de mantenimiento.

Sin embargo, también es importante considerar el mayor desgaste, menor eficiencia de combustible y características de seguridad limitadas en comparación con los modelos más nuevos, así como la posible necesidad de más mantenimiento y reparaciones a medida que el vehículo envejece.

```python
#Vamos a contar las diferentes clases de automoviles
def contar(n, vec):
  '''Para esta funcion n es el numero a contar y vec es la lista donde contar '''
  count=0
  for i in vec:
    if i==n:
      count+=1
  return count
```

```python

print('AUTOMOVIL aparece: ',contar('AUTOMOVIL', lista_clase))
print('CAMPERO aparece: ',contar('CAMPERO', lista_clase))
print('CAMIONETA aparece: ',contar('CAMIONETA', lista_clase))
print('CAMION aparece: ',contar('CAMION', lista_clase))
print('MOTOCICLETA aparece: ',contar('MOTOCICLETA', lista_clase))
print('VOLQUETA aparece: ',contar('VOLQUETA', lista_clase))
print('BUSETA aparece: ',contar('BUSETA', lista_clase))
print('MICRO BUS aparece: ',contar('MICRO BUS', lista_clase))
print('MOTOCARRO aparece: ',contar('MOTOCARRO', lista_clase))
print('CUATRIMOTO aparece: ',contar('CUATRIMOTO', lista_clase))
print('BUS aparece: ',contar('BUS', lista_clase))

aux=[contar('AUTOMOVIL', lista_clase), contar('CAMPERO', lista_clase),contar('CAMIONETA', lista_clase),
     contar('CAMION', lista_clase),contar('MOTOCICLETA', lista_clase), contar('VOLQUETA', lista_clase),
     contar('BUSETA', lista_clase) , contar('MICRO BUS', lista_clase),contar('MOTOCARRO', lista_clase),contar('CUATRIMOTO', lista_clase),contar('BUS', lista_clase)]
print('La suma total es de: ', sum(aux))
```

```python
num_clase=[]
for i in range(len(lista_clase)):
  num_clase.append(contar(lista_clase[i],lista_clase))

print(num_clase)
print(len(num_clase))
```
